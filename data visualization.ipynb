```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
sportify =pd.read_csv('./spotify - spotify.csv')

sportify.set_index('Date')

#sportify
```

```python
sportify.plot(kind='line', x='Date')
plt.title('My First visualization in pandas')

plt.show()
```

```python
sportify.plot(kind='bar')

plt.show()
```

```python
sportify.plot(kind='hist')

plt.show()
```
