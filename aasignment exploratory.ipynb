```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import opendatasets as od
%matplotlib inline
plt.style.use('ggplot')
```

```python
footballWages = pd.read_csv('./football-wages-prediction/SalaryPrediction.csv')

footballWages
```

```python
footballWages.shape
```

```python
footballWages['Wage'] = footballWages['Wage'].astype(str)
footballWages['Wage'] = footballWages['Wage'].str.split(',')
footballWages['Wage'] = footballWages['Wage'].apply(lambda x: ''.join(x))

footballWages
```

```python
footballWages['Wage'] = pd.to_numeric(footballWages['Wage'])
pd.set_option('display.float_format', '{:.2f}'.format)
footballWages.describe()
```

```python
plt.figure(figsize=(14,6))
sns.scatterplot(x=footballWages['Age'],y=footballWages['Wage'], hue=footballWages['Position']);
```

```python
plt.figure(figsize=(14,6))
sns.barplot(x=footballWages['Age'], y=footballWages['Wage']);
```
