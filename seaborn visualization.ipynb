```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
%matplotlib inline
       
```

```python
yield_apples = [0.895, 0.91, 0.919, 0.926, 0.929, 0.931]
years = [2010, 2011, 2012, 2013, 2014, 2015]
```

```python
plt.plot(years, yield_apples)
plt.xlabel('years')
plt.xlabel('yield (tonsnper hectare)');
```

```python
years = range(2000, 2012)
yield_apples = [0.895, 0.91, 0.919, 0.926, 0.929, 0.931, 0.934, 0.936, 0.937, 0.9375, 0.9372, 0.939]
yield_banana = [0.962, 0.941, 0.930, 0.923, 0.918, 0.908, 0.907, 0.904, 0.901, 0.898, 0.9, 0.896, ]
```

```python
plt.figure(figsize=(12, 6))

sns.set_style("whitegrid")

plt.plot(years, yield_apples, marker='o', color='green')
plt.plot(years, yield_banana, marker='o', color='yellow')

plt.title("Crop Yields in Kano")
plt.legend(['Apples', 'Banana']);


plt.xlabel('years')
plt.xlabel('yield');
```
