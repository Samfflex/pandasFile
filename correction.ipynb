```python
#Grouping And Aggregation
product_sales = df.groupby('PRODUCTLINE')['REVENUE'].sum()

product_sales

top_products = product_sales.sort_values(ascending=False)

top_3_products = top_products
top_3_products.iloc[0:3]

# top_3_products
```

```python
# Sort sales_df by the "Date" column in ascending order.

df['ORDERDATE'] = pd.to_datetime(df['ORDERDATE'])
order_level = df.sort_values(by='ORDERDATE', ascending=True)
# pd.to
order_level
```

```python
# Display the rows with the 5 lowest revenue values.
lowest_sale = df.sort_values(by='REVENUE', ascending=False)

lowest_sale.tail(5)
```

```python
print('The mean of the Revenue is', df['REVENUE'].mean())
print('The median of the Revenue is', df['REVENUE'].median())
print('The standard deviation of the Revenue is', df['REVENUE'].std())

```

```python
df['Discounted_Price'] = df['PRICEEACH'] * 0.90

df.head(10)
```

```python
def apply_discount(price):
    return price * 0.80

df['Omo_Iya_Discount'] = df['PRICEEACH'].map(apply_discount)
df.head(10)
```

```python
df['Omo_Ilu_Discount'] = df['PRICEEACH'].map(lambda x: x * 0.85)
df.head(10)
```

```python
df['TOTAL_SALES'] = df['PRICEEACH'] * df['QUANTITYORDERED']

df.head()
```
